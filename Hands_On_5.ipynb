<h1>There are 3 models that spacy provides for english: en_core_web_sm, en_core_web_md and en_core_web_lg. sm, md, lg stand for small, medium and large respectively.</h1>
<h2> en_core_web_sm: English multi-task CNN trained on OntoNotes <br>en_core_web_md: English multi-task CNN trained on OntoNotes, with GloVe vectors trained on Common Crawl.<br>en_core_web_lg: English multi-task CNN trained on OntoNotes, with GloVe vectors trained on Common Crawl.<br>
The size of the model has more to do with accuracy otherwise they do similar work.Also all these models have a fixed max size that is 1000000, in order to use a larger text we need to change the max size.</h2>

# Imports

```
import spacy
from urllib import request
```

# Scrapping Project Gutenberg

```
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
chunk = response.read().decode('utf8')
chunk1 = chunk[:1000000]
```

# Loading Models and Creatinglanguage models for Gutenberg

```
m = spacy.load('en_core_web_sm')
m1 = spacy.load('en_core_web_md')
m2 = spacy.load('en_core_web_lg')
```

```
doc = m(chunk1)
doc1 = m1(chunk1)
doc2 = m2(chunk1)
```

```
for ent in doc.ents:
  print(ent.text,ent.label_)
```

```
for ent in doc1.ents:
  print(ent.text,ent.label_)
```

```
for ent in doc2.ents:
  print(ent.text,ent.label_)
```
