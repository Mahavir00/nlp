# Frequency Distributions

```python
from nltk.probability import ConditionalFreqDist
```

```python
txt = inaugural.words(fileids ='2017-Trump.txt')[0:100]
```

```python
f_dict = nltk.FreqDist(txt)
f_dict
```

```python
cf_dict = ConditionalFreqDist((len(word), word) for word in txt)
cf_dict
```

### WORD CLOUD

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
stopwords = set(STOPWORDS)
```

```python
txt = inaugural.words(fileids ='2017-Trump.txt')
wordcloud = WordCloud(width = 400, height = 400, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(' '.join(list(txt))) 
```

```python
plt.figure(figsize = (6, 6), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```
